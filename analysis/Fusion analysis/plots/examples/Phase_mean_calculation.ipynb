```python
def createFrame(index,key,timeerr_dom,timeerr_nonDom,Ferr_dom,Ferr_nonDom):
    numTrialsAvg = 5
    first5 = [np.mean(timeerr_dom[key][:int(numTrialsAvg)]),np.mean(timeerr_nonDom[key][:int(numTrialsAvg)]),
              np.mean(Ferr_dom[key][:int(numTrialsAvg)]),np.mean(Ferr_nonDom[key][:int(numTrialsAvg)])]
    last5 = [np.mean(timeerr_dom[key][-int(numTrialsAvg):]),np.mean(timeerr_nonDom[key][-int(numTrialsAvg):]),
              np.mean(Ferr_dom[key][-int(numTrialsAvg):]),np.mean(Ferr_nonDom[key][-int(numTrialsAvg):])]
    average = [np.mean(timeerr_dom[key]),np.mean(timeerr_nonDom[key]),
              np.mean(Ferr_dom[key]),np.mean(Ferr_nonDom[key])]
    frame = pd.DataFrame({'ID':np.squeeze(np.matlib.repmat(key,4,1)),
                          'hand': ['Dominant','Non-Dominant','Dominant','Non-Dominant'],
                          'ErrorType': ['Time','Time','Frequency','Frequency'],
#                           'trial order': trial_order[index,:],
                          'first5': [abs(x) for x in first5],
                          'last5': [abs(x) for x in last5],
                          'average': [abs(x) for x in average]})
    return frame

# PLOT NON-DOM TO DOM
dataframe = {}
for index,key in enumerate(keys):
    dataframe[key] = createFrame(index,key,timeerr_dom,timeerr_nonDom,Ferr_dom,Ferr_nonDom)
HCPS_df = dataframe[keys[0]]
for key in keys[1:]:
    HCPS_df = pd.concat([HCPS_df,dataframe[key]])
firstlast_df = pd.melt(HCPS_df[HCPS_df['ErrorType']=='Time'],id_vars=['ID','hand','ErrorType'],value_vars=['first5','last5'])
NDtoD_time = firstlast_df
def plot_boxplot_timeerr(firstlast_df,ylim,ylabel):
    plt.figure(figsize=(7,5))
    ax = sns.boxplot(x='hand',y='value',hue='variable',data=firstlast_df,order=['Non-Dominant','Dominant'])
    # ax.set_title('Time-domain error - did people learn?')
    ax.set_ylabel(ylabel)
    ax.set_ylim(ylim)
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.tick_params(length=10,width=3,direction='inout')
    plt.legend(frameon=False)
    plt.show()

    # run statistical tests
    # 2 way anova
    table = FF.create_table(firstlast_df.drop(['ErrorType','ID'],axis=1))
    formula = 'value ~ C(hand) + C(variable) + C(hand):C(variable)'
    model = ols(formula, firstlast_df.drop(['ErrorType','ID'],axis=1)).fit()
    aov_table = statsmodels.stats.anova.anova_lm(model, typ=2)
    print(aov_table)

    # post-hoc tests
    print('Time domain results:')
    groups = ['Dominant','Non-Dominant']
    for group in groups:
        firstlast_dfgrouped = firstlast_df.groupby('hand').get_group(group)
        print(group,': ')
        print(wilcoxon(*[group['value'].values for name, group in firstlast_dfgrouped.groupby('variable')]))   
        print(ttest_rel(*[group['value'].values for name, group in firstlast_dfgrouped.groupby('variable')]))
        print(shapiro(firstlast_dfgrouped[firstlast_dfgrouped['variable']=='first5']['value']))
        print(shapiro(firstlast_dfgrouped[firstlast_dfgrouped['variable']=='last5']['value']))
        print('')
    print('non-dom statistically significant (people learned) but dom not statistically significant (people did not learn)')

    groups = ['first5','last5']
    for group in groups:
        firstlast_dfgrouped = firstlast_df.groupby('variable').get_group(group)
        print(group,': ')
        print(wilcoxon(*[group['value'].values for name, group in firstlast_dfgrouped.groupby('hand')]))
        print(ttest_rel(*[group['value'].values for name, group in firstlast_dfgrouped.groupby('hand')]))
        print(shapiro(firstlast_dfgrouped[firstlast_dfgrouped['hand']=='Dominant']['value']))
        print(shapiro(firstlast_dfgrouped[firstlast_dfgrouped['hand']=='Non-Dominant']['value']))
        print('')
    print('non-dominant hand had better first 5 results than dominant hand but same last5 results')

    firstlast_dfgrouped = firstlast_df.groupby('hand').get_group('Non-Dominant')
    nonDom_wilcoxon = firstlast_dfgrouped[firstlast_dfgrouped['variable']=='last5']['value'].values
    firstlast_dfgrouped = firstlast_df.groupby('hand').get_group('Dominant')
    dom_wilcoxon = firstlast_dfgrouped[firstlast_dfgrouped['variable']=='first5']['value'].values
    print('Transfer NDtoD: ',wilcoxon(nonDom_wilcoxon,dom_wilcoxon))

plot_boxplot_timeerr(firstlast_df,ylim=(0,.8),ylabel='$MSE_{time}$')
    
firstlastFF_df = pd.melt(HCPS_df[HCPS_df['ErrorType']=='Frequency'],id_vars=['ID','hand','ErrorType'],value_vars=['first5','last5'])

plt.figure(figsize=(7,5))
ax = sns.boxplot(x='hand',y='value',hue='variable',data=firstlastFF_df,order=['Non-Dominant','Dominant'])
# ax.set_title('Frequency-domain error - did people learn?')
# ax.set_ylim((0,30))
ax.set_ylabel('$MSE_{freq}$')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.tick_params(length=10,width=3,direction='inout')
plt.legend(frameon=False)
plt.show()

# run statistical tests


print('\nFrequency domain results:')
groups = ['Dominant','Non-Dominant']
for group in groups:
    firstlastFF_dfgrouped = firstlastFF_df.groupby('hand').get_group(group)
    print(group,': ')
    print(wilcoxon(*[group['value'].values for name, group in firstlastFF_dfgrouped.groupby('variable')]))
    print(ttest_rel(*[group['value'].values for name, group in firstlastFF_dfgrouped.groupby('variable')]))
    print('')
print('neither groups learned feedforward controller?')

#  PLOT DOM TO NON-DOM
dataframe = {}
for index,key in enumerate(keys2):
    dataframe[key] = createFrame(index,key,timeerr_dom,timeerr_nonDom,Ferr_dom,Ferr_nonDom)
HCPS_df = dataframe[keys2[0]]
for key in keys2[1:]:
    HCPS_df = pd.concat([HCPS_df,dataframe[key]])
firstlast_df = pd.melt(HCPS_df[HCPS_df['ErrorType']=='Time'],id_vars=['ID','hand','ErrorType'],value_vars=['first5','last5'])
DtoND_time = firstlast_df
plt.figure(figsize=(7,5))
ax = sns.boxplot(x='hand',y='value',hue='variable',data=firstlast_df)
# ax.set_title('Time-domain error - did people learn?')
ax.set_ylabel('$MSE_{time}$')
ax.set_ylim((0,.8))
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.tick_params(length=10,width=3,direction='inout')
plt.legend(frameon=False)

firstlastFF_df = pd.melt(HCPS_df[HCPS_df['ErrorType']=='Frequency'],id_vars=['ID','hand','ErrorType'],value_vars=['first5','last5'])

plt.figure(figsize=(7,5))
ax = sns.boxplot(x='hand',y='value',hue='variable',data=firstlastFF_df)
# ax.set_title('Frequency-domain error - did people learn?')
ax.set_ylabel('$MSE_{freq}$')
# ax.set_ylim((0,30))
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.tick_params(length=10,width=3,direction='inout')
plt.legend(frameon=False)
plt.show()

# run statistical tests

# ANOVA
table = FF.create_table(firstlast_df.drop('ErrorType',axis=1))
formula = 'value ~ C(hand) + C(variable) + C(hand):C(variable)'
model = ols(formula, firstlast_df.drop('ErrorType',axis=1)).fit()
aov_table = statsmodels.stats.anova.anova_lm(model, typ=2)
print(aov_table)

firstlastFF_df['value'] = abs(firstlastFF_df['value'])
table = FF.create_table(firstlastFF_df.drop('ErrorType',axis=1))
formula = 'value ~ C(hand) + C(variable) + C(hand):C(variable)'
model = ols(formula, firstlastFF_df.drop('ErrorType',axis=1)).fit()
aov_table = statsmodels.stats.anova.anova_lm(model, typ=2)
print(firstlastFF_df.drop('ErrorType',axis=1).describe())
print(aov_table)

# post-hoc tests
print('Time domain results:')
groups = ['Dominant','Non-Dominant']
for group in groups:
    firstlast_dfgrouped = firstlast_df.groupby('hand').get_group(group)
    print(group,': ')
    print(wilcoxon(*[group['value'].values for name, group in firstlast_dfgrouped.groupby('variable')]))
    print(ttest_rel(*[group['value'].values for name, group in firstlast_dfgrouped.groupby('variable')]))
    print('')
print('non-dom statistically significant (people learned) but dom not statistically significant (people did not learn)')
firstlast_dfgrouped = firstlast_df.groupby('hand').get_group('Non-Dominant')
nonDom_wilcoxon = firstlast_dfgrouped[firstlast_dfgrouped['variable']=='first5']['value'].values
firstlast_dfgrouped = firstlast_df.groupby('hand').get_group('Dominant')
dom_wilcoxon = firstlast_dfgrouped[firstlast_dfgrouped['variable']=='last5']['value'].values
print('Transfer NDtoD: ',wilcoxon(nonDom_wilcoxon,dom_wilcoxon))

print('\nFrequency domain results:')
groups = ['Dominant','Non-Dominant']
for group in groups:
    firstlastFF_dfgrouped = firstlastFF_df.groupby('hand').get_group(group)
    print(group,': ')
    print(wilcoxon(*[group['value'].values for name, group in firstlastFF_dfgrouped.groupby('variable')]))
    print(ttest_rel(*[group['value'].values for name, group in firstlastFF_dfgrouped.groupby('variable')]))
    print('')
print('neither groups learned feedforward controller?')

groups = ['first5','last5']
for group in groups:
    firstlast_dfgrouped = firstlast_df.groupby('variable').get_group(group)
    print(group,': ')
    print(wilcoxon(*[group['value'].values for name, group in firstlast_dfgrouped.groupby('hand')]))
    print(ttest_rel(*[group['value'].values for name, group in firstlast_dfgrouped.groupby('hand')]))
    print(shapiro(firstlast_dfgrouped[firstlast_dfgrouped['hand']=='Dominant']['value']))
    print(shapiro(firstlast_dfgrouped[firstlast_dfgrouped['hand']=='Non-Dominant']['value']))
    print('')
print('non-dominant hand had better first 5 results than dominant hand but same last5 results')

```

```python
def extract_Yerr(error,keys,ff=2):
    primes = np.asarray([2, 3, 5, 7, 11, 13, 17, 19])# max =37
    IX = primes*2
    data = []
    for index, key in enumerate(keys):
        trials = list(error[key].keys())
        err = []
        for trial in trials:
            err.append(np.linalg.norm(error[key][trial]['OUTS'][IX][:ff] - error[key][trial]['REFS'][IX][:ff],ord=2,axis=0))
        data.append(err)
    data = np.asarray(data)
    return data

def get_quartiles_MSE(error,keys):
    data = extract_Yerr(error,keys,ff=2)
    return np.percentile(data,[75,50,25],axis=0)
    
dom_FREQ = get_quartiles_MSE(FREQVALS_dom,keys)
dom2_FREQ = get_quartiles_MSE(FREQVALS_dom,keys2)
nonDom_FREQ = get_quartiles_MSE(FREQVALS_nonDom,keys)
nonDom2_FREQ = get_quartiles_MSE(FREQVALS_nonDom,keys2)

connected_fig(dom_FREQ,dom2_FREQ,nonDom_FREQ,nonDom2_FREQ)


def createFrame(dom,nonDom,keys,keys2):
    ff = 2 # number of frequencies to use
    ND2D_dom = extract_Yerr(dom,keys,ff=ff) # 9 people x 30 trials
    ND2D_nonDom = extract_Yerr(nonDom,keys,ff=ff)
    D2ND_dom = extract_Yerr(dom,keys2,ff=ff)
    D2ND_nonDom = extract_Yerr(nonDom,keys2,ff=ff)
    numTrialsAvg = 5
    first5 = [np.mean(D2ND_dom[:,:int(numTrialsAvg)],axis=1),np.mean(D2ND_nonDom[:,:int(numTrialsAvg)],axis=1),
              np.mean(ND2D_nonDom[:,:int(numTrialsAvg)],axis=1),np.mean(ND2D_dom[:,:int(numTrialsAvg)],axis=1)]
    last5 = [np.mean(D2ND_dom[:,-int(numTrialsAvg):],axis=1),np.mean(D2ND_nonDom[:,-int(numTrialsAvg):],axis=1),
              np.mean(ND2D_nonDom[:,-int(numTrialsAvg):],axis=1),np.mean(ND2D_dom[:,-int(numTrialsAvg):],axis=1)]
    frame = pd.DataFrame({'ID':np.asarray([keys2,keys2,keys,keys]).flatten(),
                          'hand': np.matlib.repmat(['Dominant','Non-Dominant','Non-Dominant','Dominant'],9,1).T.flatten(),
                          'ErrorType': np.matlib.repmat(['D2ND','D2ND','ND2D','ND2D'],9,1).T.flatten(),
#                           'trial order': trial_order[index,:],
                          'first5': np.asarray(first5).flatten(),
                          'last5': np.asarray(last5).flatten()})
    return frame
dataframe = {}
dataframe = createFrame(FREQVALS_dom,FREQVALS_nonDom,keys,keys2)
firstlast_df = pd.melt(dataframe[dataframe['ErrorType']=='D2ND'],id_vars=['ID','hand','ErrorType'],value_vars=['first5','last5'])

plot_boxplot_timeerr(firstlast_df,ylim=(0,1),ylabel='$|R-Y|_2$')

firstlast_df = pd.melt(dataframe[dataframe['ErrorType']=='ND2D'],id_vars=['ID','hand','ErrorType'],value_vars=['first5','last5'])

plot_boxplot_timeerr(firstlast_df,ylim=(0,1),ylabel='$|R-Y|_2$')
# plot_boxplots(FREQVALS_dom,FREQVALS_nonDom,[keys,keys2],createFrame,'$|H_{yr}-1|_2$',(0,3))
```

```python
# to get magnitude IQR
def get_FA_mag(FA_data, keys,ff):
    data = []
    for index, key in enumerate(keys):
        FA = FA_data[key][:,ff]
        data.append(FA)
    data = np.asarray(data) # 9 users x 29 trials

    # obtain IQR of magnitude
    data_mag = np.abs(data)
    data_magq25,data_magq50,data_magq75 = np.percentile(data_mag,[25,50,75],axis=0)
    return data_magq25,data_magq50,data_magq75

# obtain IQR of phase
def get_FA_angle(FA_data, keys,ff):
    polar2z = lambda theta: 1*np.exp(1j*theta)
    data = []
    for index, key in enumerate(keys):
        FA = FA_data[key][:,ff]
        data.append(FA)
    data = np.asarray(data) # 9 users x 29 trials

    # obtain IQR of angle
    data_angle = np.angle(data) # get angle of each datapoint
    data_unit = polar2z(data_angle) # get the unit vectors of each angle 
    mean_angle = np.mean(data_unit,axis=0) # find mean unit vector
    data_centered_unit = np.divide(data_unit, np.tile(mean_angle,(9,1))) # center each datapoint about mean by subtracting mean
    data_centered_angle = np.angle(data_centered_unit) # now make back in angles instead of unit vectors
    data_percentiles_centered = np.percentile(data_centered_angle,[25,50,75],axis=0) # find percentiles
    data_percentiles = data_percentiles_centered + np.tile(np.angle(mean_angle),(3,1)) # add angle back in
#     data_angleq25,data_angleqq50,data_angleqq75 = data_percentiles
    return  data_percentiles #data_angleq25,data_angleqq50,data_angleqq75

def FA_make_plots(dom_FA,dom2_FA,nonDom_FA,nonDom2_FA):
    dom_FAq25,dom_FAq50,dom_FAq75 = dom_FA
    dom2_FAq25,dom2_FAq50,dom2_FAq75 = dom2_FA
    nonDom_FAq25,nonDom_FAq50,nonDom_FAq75 = nonDom_FA
    nonDom2_FAq25,nonDom2_FAq50,nonDom2_FAq75 = nonDom2_FA
    # fit log function to median
    xlog = np.linspace(1,29,1000)
    xx = np.linspace(1,29,29)
    dom_FA_log = np.polyfit(np.log(xx),dom_FAq50,1)
    dom2_FA_log = np.polyfit(np.log(xx),dom2_FAq50,1)
    nonDom_FA_log = np.polyfit(np.log(xx),nonDom_FAq50,1)
    nonDom2_FA_log = np.polyfit(np.log(xx),nonDom2_FAq50,1)
    f,axs = plt.subplots(2,2,sharex=True,sharey=True, figsize=(20,10))
    axs[0,0].errorbar(x=xx,y=dom2_FAq50,yerr=np.asarray([dom2_FAq50-dom2_FAq25,dom2_FAq75-dom2_FAq50]),xerr=None,
                 marker='o',markersize=markersize,color=sns.color_palette()[1],fmt='none',
                 linewidth=linewidth,label='DtoND-D')
    axs[0,0].scatter(xx,dom2_FAq50,marker='o',s=markersize,color=sns.color_palette()[1]) # plot median as data point
    axs[0,0].plot(xlog,dom2_FA_log[0]*np.log(xlog)+dom2_FA_log[1],linewidth=linewidth,color=sns.color_palette()[0])
    axs[0,0].set_ylabel('FF')
    axs[0,0].spines['top'].set_visible(False)
    axs[0,0].spines['right'].set_visible(False)

    axs[0,1].errorbar(x=xx,y=nonDom2_FAq50,yerr=np.asarray([nonDom2_FAq50-nonDom2_FAq25,nonDom2_FAq75-nonDom2_FAq50]),xerr=None,
                 marker='o',markersize=markersize,linewidth=linewidth,color=sns.color_palette()[1],label='DtoND-ND',fmt='none')
    axs[0,1].scatter(xx,nonDom2_FAq50,marker='o',s=markersize,color=sns.color_palette()[1])
    axs[0,1].plot(xlog,nonDom2_FA_log[0]*np.log(xlog)+nonDom2_FA_log[1],linewidth=linewidth,color=sns.color_palette()[0])
    axs[0,1].spines['top'].set_visible(False)
    axs[0,1].spines['right'].set_visible(False)

    axs[1,0].errorbar(x=xx,y=dom_FAq50,yerr=np.asarray([dom_FAq50-dom_FAq25,dom_FAq75-dom_FAq50]),xerr=None,
                 marker='o',markersize=markersize,linewidth=linewidth,color=sns.color_palette()[1],label='NDtoD-D',fmt='none')
    axs[1,0].scatter(xx,dom_FAq50,marker='o',s=markersize,color=sns.color_palette()[1])
    axs[1,0].plot(xlog,dom_FA_log[0]*np.log(xlog)+dom_FA_log[1],linewidth=linewidth,color=sns.color_palette()[0])
    axs[1,0].set_xlabel('Trials')
    axs[1,0].set_ylabel('FF')
    axs[1,0].spines['top'].set_visible(False)
    axs[1,0].spines['right'].set_visible(False)

    axs[1,1].errorbar(x=xx,y=nonDom_FAq50,yerr=np.asarray([nonDom_FAq50-nonDom_FAq25,nonDom_FAq75-nonDom_FAq50]),xerr=None,
                 marker='o',markersize=markersize,linewidth=linewidth,color=sns.color_palette()[1],label='NDtoD-ND',fmt='none')
    axs[1,1].scatter(xx,nonDom_FAq50,marker='o',s=markersize,color=sns.color_palette()[1])
    axs[1,1].plot(xlog,nonDom_FA_log[0]*np.log(xlog)+nonDom_FA_log[1],linewidth=linewidth,color=sns.color_palette()[0])
    axs[1,1].set_xlabel('Trials')
    axs[1,1].spines['top'].set_visible(False)
    axs[1,1].spines['right'].set_visible(False)
    plt.tight_layout()
    plt.show()

# get IQR of magnitude and plot
print('magnitude')
dom_FAmag = get_FA_mag(FA_dom,keys,ff=0) # NDtoD-D
dom2_FAmag = get_FA_mag(FA_dom,keys2,ff=0) # DtoND-D
nonDom_FAmag = get_FA_mag(FA_nonDom,keys,ff=0) # NDtoD-ND
nonDom2_FAmag = get_FA_mag(FA_nonDom,keys2,ff=0) # DtoND-ND
FA_make_plots(dom_FAmag,dom2_FAmag,nonDom_FAmag,nonDom2_FAmag)

# get IQR of angle and plot
print('angle')
dom_FAangle = get_FA_angle(FA_dom,keys,ff=0) # NDtoD-D
dom2_FAangle = get_FA_angle(FA_dom,keys2,ff=0) # DtoND-D
nonDom_FAangle = get_FA_angle(FA_nonDom,keys,ff=0) # NDtoD-ND
nonDom2_FAangle = get_FA_angle(FA_nonDom,keys2,ff=0) # DtoND-ND
FA_make_plots(dom_FAangle,dom2_FAangle,nonDom_FAangle,nonDom2_FAangle)
```

```python
# to get magnitude IQR
def get_mag(FA_data, keys,ff):
    data = []
    for index, key in enumerate(keys):
        FA = FA_data[key][:,:ff]
        data.append(FA)
    data = np.asarray(data) # 9 users x 29 trials

    # obtain IQR of magnitude
    data_mag = np.abs(data)
    data_magq25,data_magq50,data_magq75 = np.percentile(data_mag,[25,50,75],axis=0)
    return data_magq25,data_magq50,data_magq75

# obtain IQR of phase
def get_angle(FA_data, keys,ff):
    polar2z = lambda theta: 1*np.exp(1j*theta)
    data = []
    for index, key in enumerate(keys):
        FA = FA_data[key][:,:ff]
        data.append(FA)
    data = np.asarray(data) # 9 users x 29 trials

    # obtain IQR of angle
    data_angle = np.angle(data) # get angle of each datapoint
    data_unit = polar2z(data_angle) # get the unit vectors of each angle 
    mean_angle = np.mean(data_unit,axis=0) # find mean unit vector
    data_centered_unit = np.divide(data_unit, np.tile(mean_angle,(9,1,1))) # center each datapoint about mean by subtracting mean
    data_centered_angle = np.angle(data_centered_unit) # now make back in angles instead of unit vectors
    data_percentiles_centered = np.percentile(data_centered_angle,[25,50,75],axis=0) # find percentiles
    data_percentiles = data_percentiles_centered + np.tile(np.angle(mean_angle),(3,1,1)) # add angle back in
#     data_angleq25,data_angleqq50,data_angleqq75 = data_percentiles
    return  data_percentiles #data_angleq25,data_angleqq50,data_angleqq75

def make_bode_plot(dom,nonDom,keys,keys2,get_mag,get_angle,ff=8):
    
    xx = [0.1,0.15,0.25,0.35,0.55,0.65,0.85,0.95]
    dom_mag = get_mag(dom,keys,ff=ff) # NDtoD-D
    dom2_mag = get_mag(dom,keys2,ff=ff) # DtoND-D
    nonDom_mag = get_mag(nonDom,keys,ff=ff) # NDtoD-ND
    nonDom2_mag = get_mag(nonDom,keys2,ff=ff) # DtoND-ND # 25th, 50th, 75th, 29 trials x 8 frequencies
    
    dom_angle = get_angle(dom,keys,ff=ff) # NDtoD-D
    dom2_angle = get_angle(dom,keys2,ff=ff) # DtoND-D
    nonDom_angle = get_angle(nonDom,keys,ff=ff) # NDtoD-ND
    nonDom2_angle = get_angle(nonDom,keys2,ff=ff) # DtoND-ND
    
    f,axs = plt.subplots(2,2,sharex=True, figsize=(20,10))
    
    axs[0,0].set_xscale("log", nonposy='clip')
    axs[0,0].spines['right'].set_visible(False)
    axs[0,0].spines['top'].set_visible(False)
    axs[0,0].fill_between(xx,np.mean(nonDom2_mag[0][-5:,:],axis=0),
                     np.mean(nonDom2_mag[2][-5:,:],axis=0), color=(.6,.6,1),alpha=.8,label='Right')#"#3F5D7D")
    axs[0,0].set_xscale("log", nonposy='clip')
    axs[0,0].spines['right'].set_visible(False)
    axs[0,0].spines['top'].set_visible(False)
#     axs[0,0].fill_between(xx,np.mean(nonDom2_mag[0][-5:,:],axis=0),
#                      np.mean(nonDom2_mag[2][-5:,:],axis=0), color=(.6,.6,1),alpha=.3,label='Left')#"#3F5D7D")
    axs[0,0].fill_between(xx,np.mean(nonDom2_mag[0][:5,:],axis=0),
                     np.mean(nonDom2_mag[2][:5,:],axis=0), color=(.6,.6,1),alpha=.3,label='Left')#"#3F5D7D")
    
    
    axs[1,0].set_xscale("log", nonposy='clip')
    axs[1,0].spines['right'].set_visible(False)
    axs[1,0].spines['top'].set_visible(False)
    axs[1,0].fill_between(xx,np.mean(nonDom2_angle[0][-5:,:],axis=0),
                     np.mean(nonDom2_angle[2][-5:,:],axis=0), color=(.6,.6,1),alpha=.8,label='Right')#"#3F5D7D")
    axs[1,0].set_xscale("log", nonposy='clip')
    axs[1,0].spines['right'].set_visible(False)
    axs[1,0].spines['top'].set_visible(False)
#     axs[1,0].fill_between(xx,np.mean(nonDom_angle[0][-5:,:],axis=0),
#                      np.mean(nonDom_angle[2][-5:,:],axis=0), color=(.3,.3,1),alpha=.3,label='Left')#"#3F5D7D")
    axs[1,0].fill_between(xx,np.mean(nonDom2_angle[0][:5,:],axis=0),
                     np.mean(nonDom2_angle[2][:5,:],axis=0), color=(.3,.3,1),alpha=.3,label='Left')#"#3F5D7D")
    
    axs[0,1].set_xscale("log", nonposy='clip')
    axs[0,1].spines['right'].set_visible(False)
    axs[0,1].spines['top'].set_visible(False)
#     axs[0,1].fill_between(xx,np.mean(dom_mag[0][-5:,:],axis=0),
#                      np.mean(dom_mag[2][-5:,:],axis=0), color=(.6,.6,1),alpha=.8,label='Right')#"#3F5D7D")
    axs[0,1].fill_between(xx,np.mean(nonDom_mag[0][:5,:],axis=0),
                     np.mean(nonDom_mag[2][:5,:],axis=0), color=(.6,.6,1),alpha=.8,label='Right')#"#3F5D7D")
    axs[0,1].set_xscale("log", nonposy='clip')
    axs[0,1].spines['right'].set_visible(False)
    axs[0,1].spines['top'].set_visible(False)
    axs[0,1].fill_between(xx,np.mean(nonDom_mag[0][-5:,:],axis=0),
                     np.mean(nonDom_mag[2][-5:,:],axis=0), color=(.6,.6,1),alpha=.3,label='Left')#"#3F5D7D")
    
    
    axs[1,1].set_xscale("log", nonposy='clip')
    axs[1,1].spines['right'].set_visible(False)
    axs[1,1].spines['top'].set_visible(False)
#     axs[1,1].fill_between(xx,np.mean(dom_angle[0][-5:,:],axis=0),
#                      np.mean(dom_angle[2][-5:,:],axis=0), color=(.6,.6,1),alpha=.8,label='Right')#"#3F5D7D")
    axs[1,1].fill_between(xx,np.mean(nonDom_angle[0][:5,:],axis=0),
                     np.mean(nonDom_angle[2][:5,:],axis=0), color=(.6,.6,1),alpha=.8,label='Right')#"#3F5D7D")
    axs[1,1].set_xscale("log", nonposy='clip')
    axs[1,1].spines['right'].set_visible(False)
    axs[1,1].spines['top'].set_visible(False)
    axs[1,1].fill_between(xx,np.mean(nonDom_angle[0][-5:,:],axis=0),
                     np.mean(nonDom_angle[2][-5:,:],axis=0), color=(.3,.3,1),alpha=.3,label='Left')#"#3F5D7D")
make_bode_plot(Hdy_dom,Hdy_nonDom,keys,keys2,get_mag,get_angle)
print('Hyd')
plt.show()
make_bode_plot(Hry_dom,Hry_nonDom,keys,keys2,get_mag,get_angle)
print('Hyr')
plt.show()


data = [Hry_dom,Hry_nonDom,Hdy_dom,Hdy_nonDom,keys,keys2]
with open('HyrHyd.pkl', 'wb') as outfile:
    pickle.dump(data, outfile, pickle.HIGHEST_PROTOCOL)
```

```python
def compute_2norm_delta(dom,nonDom,keys,numTrialsAvg=5,ff=2):
    data_dom = []
    for index, key in enumerate(keys):
        FA = dom[key][:,:ff]#-np.tile(np.average(dom[key][-5:,:ff],axis=0),(29,1))
        data_dom.append(FA)
    data_dom = np.asarray(data_dom) # 9 users x 29 trials x 8 frequencies
    data_dom_norm = np.linalg.norm(data_dom,ord=2,axis=2)
    last5=np.mean(data_dom_norm[:,-5:],axis=1)
    first5=np.mean(data_dom_norm[:,:5],axis=1)
    data_dom_perchange = (last5-first5)/first5

    data_nonDom = []
    for index, key in enumerate(keys):
        FA = nonDom[key][:,:ff]#-np.tile(np.average(nonDom[key][-5:,:ff],axis=0),(29,1))
        data_nonDom.append(FA)
    data_nonDom = np.asarray(data_nonDom) # 9 users x 29 trials x 8 frequencies
    data_nonDom_norm = np.linalg.norm(data_nonDom,ord=2,axis=2) # 9 users x 29 trials
    last5=np.mean(data_nonDom_norm[:,-5:],axis=1)
    first5=np.mean(data_nonDom_norm[:,:5],axis=1)
    data_nonDom_perchange = (last5-first5)/first5
    
    return data_dom_perchange, data_nonDom_perchange

def plot_deltaHF(Hyr_dom,Hyr_nonDom,F_dom,F_nonDom,keys,title,compute_2norm_delta=compute_2norm_delta):
    Hyr_dom_perchange,Hyr_nonDom_perchange = compute_2norm_delta(Hyr_dom,Hyr_nonDom,keys)
    F_dom_perchange,F_nonDom_perchange = compute_2norm_delta(F_dom,F_nonDom,keys)
    plt.figure(figsize=(10,5))
    plt.scatter(F_dom_perchange,Hyr_dom_perchange,c='r',label='Right hand')
    plt.scatter(F_nonDom_perchange,Hyr_nonDom_perchange,c='b',label='Left hand')
    plt.xlabel('F_percentChange')
    plt.ylabel('H_percentChange')
    plt.legend()
    plt.title(title)
    plt.show()

plot_deltaHF(Hry_dom,Hry_nonDom,FA_dom,FA_nonDom,keys,'Hyr vs F for left to right')
plot_deltaHF(Hry_dom,Hry_nonDom,FA_dom,FA_nonDom,keys2,'Hyr vs F for right to left')
```
