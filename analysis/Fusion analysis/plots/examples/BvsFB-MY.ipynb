```python
pd_frame = pd.read_pickle("BvsFB.pkl")

colors = [(.6,.6,1),(1,.6,.6)]
colors = [(0.,0.,0.),(.5,.5,.5)]
plt.figure(figsize=(7,5))
ax = sns.boxplot(x='Frequency',y='Data',hue='Model',
                 data=pd_frame,showfliers=False,palette=colors)

# boxes
plt.setp(ax.artists, edgecolor = 'w')
# [fliers,] whisker, connector, connector, whisker, median
plt.setp(ax.lines[4::5], color='w')

ax.set_ylim(0,1)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.set_ylabel('coefficient of determination ($r^2$)')

ax.set_xticklabels(ax.get_xticklabels(), **font)
ax.set_yticklabels(['%0.1f'%t for t in ax.get_yticks()], **font)



plt.xlabel('frequency (Hz)')

plt.legend(ncol=2,loc='lower center')
l = ax.get_legend()
lbls = [r'$B$', r'$F+B$']
for txt, lbl in zip(l.get_texts(), lbls): 
    txt.set_text(lbl)
l.set_title(None)

plt.tight_layout()

fig = plt.gcf()
fig.savefig('BvsFB.svg')

plt.show()
```
